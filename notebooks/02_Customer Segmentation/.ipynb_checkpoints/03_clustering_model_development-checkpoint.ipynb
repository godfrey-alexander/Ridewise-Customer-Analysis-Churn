```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA


import warnings # Ignore all warnings
warnings.filterwarnings("ignore")

# Settings
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
pd.options.display.float_format = '{:.2f}'.format
```

```python
df= pd.read_csv('../../data/new_data/data_preprocessed.csv')
df.head()
```

```python
exclude_cols = ['loyalty_status', 'city_Cairo', 'city_Lagos', 'city_Nairobi']
X = df.drop(columns=exclude_cols)
X.info()
```

```python
Why did we exclude these columns?

```

```python
# Elbow method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20, max_iter=500)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot inertia values
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
# Compute silhouette scores
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20, max_iter=500)
    kmeans.fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Plot silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title("Silhouette Scores")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
K_range = range(2, 11)
sample_n = min(10000, len(X))
inertias, sils, dbs, chs = [], [], [], []

print(f"{'K':>3} | {'Inertia':>12} | {'Silhouette':>10} | {'Davies-B':>9} | {'Calinski-H':>12}")
print("-"*60)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=20, max_iter=500)
    labels = km.fit_predict(X)

    # Sample for silhouette to keep it fast
    sil = silhouette_score(X, labels, sample_size=sample_n, random_state=42)
    db  = davies_bouldin_score(X, labels)
    ch  = calinski_harabasz_score(X, labels)

    inertias.append(km.inertia_); sils.append(sil); dbs.append(db); chs.append(ch)
    print(f"{k:>3} | {km.inertia_:>12,.0f} | {sil:>10.4f} | {db:>9.4f} | {ch:>12,.0f}")
```

```python
Ks = list(K_range)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Inertia (Elbow)
axes[0,0].plot(Ks, inertias, 'o-', color='steelblue')
axes[0,0].set_title('Elbow (Inertia)')
axes[0,0].set_xlabel('K'); axes[0,0].set_ylabel('Inertia')

# Silhouette (higher better)
axes[0,1].plot(Ks, sils, 'o-', color='green')
axes[0,1].axhline(max(sils), ls='--', color='red', alpha=0.5)
axes[0,1].set_title('Silhouette')
axes[0,1].set_xlabel('K'); axes[0,1].set_ylabel('Score')

# Davies-Bouldin (lower better)
axes[1,0].plot(Ks, dbs, 'o-', color='purple')
axes[1,0].axhline(min(dbs), ls='--', color='orange', alpha=0.5)
axes[1,0].set_title('DaviesBouldin')
axes[1,0].set_xlabel('K'); axes[1,0].set_ylabel('Score')

# CalinskiHarabasz (higher better)
axes[1,1].plot(Ks, chs, 'o-', color='darkred')
axes[1,1].axhline(max(chs), ls='--', color='gray', alpha=0.5)
axes[1,1].set_title('CalinskiHarabasz')
axes[1,1].set_xlabel('K'); axes[1,1].set_ylabel('Score')

plt.tight_layout()
plt.show()
```

```python
# Elbow (Inertia), Silhouette, DaviesBouldin, CalinskiHarabasz
```

#### Segmenting on 2 clusters

```python
k_selected = 2

sample_n = min(10000, len(X))

final_kmeans = KMeans(
    n_clusters=k_selected,
    random_state=42,
    n_init=20,
    max_iter=500
)
labels = final_kmeans.fit_predict(X)
df['Cluster'] = labels

print(f" Created {k_selected} clusters")
print("\nCluster distribution:")
print(df['Cluster'].value_counts(normalize=True).sort_index())

# Quality metrics
sil = silhouette_score(X, labels, sample_size=sample_n, random_state=42)
db  = davies_bouldin_score(X, labels)
ch  = calinski_harabasz_score(X, labels)
print(f"\nSilhouette: {sil:.4f} | DaviesBouldin: {db:.4f} | CalinskiHarabasz: {ch:,.0f}")
```

```python
# Apply PCA to reduce dimensionality to 2 components
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

# Create a DataFrame for plotting
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Cluster'] = df['Cluster']

# Visualize Clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Cluster',
    data=pca_df,
    palette='viridis',
    legend='full',
    alpha=0.7
)
plt.title('2D PCA of Clusters (K=4)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

#### Segmenting on 4 clusters

```python
k_selected = 4

sample_n = min(10000, len(X))

final_kmeans = KMeans(
    n_clusters=k_selected,
    random_state=42,
    n_init=20,
    max_iter=500
)
labels = final_kmeans.fit_predict(X)
df['Cluster'] = labels

print(f" Created {k_selected} clusters")
print("\nCluster distribution:")
print(df['Cluster'].value_counts(normalize=True).sort_index())

# Quality metrics
sil = silhouette_score(X, labels, sample_size=sample_n, random_state=42)
db  = davies_bouldin_score(X, labels)
ch  = calinski_harabasz_score(X, labels)
print(f"\nSilhouette: {sil:.4f} | DaviesBouldin: {db:.4f} | CalinskiHarabasz: {ch:,.0f}")
```

```python
# Apply PCA to reduce dimensionality to 2 components
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

# Create a DataFrame for plotting
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Cluster'] = df['Cluster']

# Visualize Clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Cluster',
    data=pca_df,
    palette='viridis',
    legend='full',
    alpha=0.7
)
plt.title('2D PCA of Clusters (K=4)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

#### Key Findings and Comparison (K=2 vs K=4)
- Silhouette Score: K=2 (0.1383) is higher than K=4 (0.1020), suggesting better-defined clusters at K=2.
- DaviesBouldin Score: K=2 (2.3036) is slightly higher than K=4 (2.2020). A lower score is better, so K=4 performs slightly better here, though both are relatively high.
- CalinskiHarabasz Score: K=2 (1,776) is higher than K=4 (1,225), suggesting better clustering at K=2.

Overall, the metrics consistently point towards K=2 as providing a slightly better or more distinct clustering solution compared to K=4, according to the Silhouette and Calinski-Harabasz scores. The Davies-Bouldin score, which ideally should be lower, showed a slightly worse performance for K=2.

These results indicate that the data might not have very strong, clearly separated clusters regardless of the number of clusters chosen, but K=2 offered a marginally better overall quality based on multiple metrics.

### Analyze the features that contribute most to the clusters

```python
k_selected = 2

sample_n = min(10000, len(X))

final_kmeans = KMeans(
    n_clusters=k_selected,
    random_state=42,
    n_init=20,
    max_iter=500
)
labels = final_kmeans.fit_predict(X)
df['Cluster'] = labels

print(f" Created {k_selected} clusters")
print("\nCluster distribution:")
print(df['Cluster'].value_counts(normalize=True).sort_index())

# Quality metrics
sil = silhouette_score(X, labels, sample_size=sample_n, random_state=42)
db  = davies_bouldin_score(X, labels)
ch  = calinski_harabasz_score(X, labels)
print(f"\nSilhouette: {sil:.4f} | DaviesBouldin: {db:.4f} | CalinskiHarabasz: {ch:,.0f}")
```

```python
cluster_centroids = df.groupby('Cluster').mean(numeric_only=True)
print(cluster_centroids)
```

```python
cluster_centroids_cleaned = cluster_centroids.drop(columns=exclude_cols)
cluster_centroids_melted = cluster_centroids_cleaned.reset_index().melt(id_vars='Cluster', var_name='Feature', value_name='Mean_Value')

print("Cleaned and Melted Cluster Centroids for Visualization:")
print(cluster_centroids_melted.head())
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(
    x='Feature',
    y='Mean_Value',
    hue='Cluster',
    data=cluster_centroids_melted,
    palette='viridis'
)
plt.title('Mean Value of Each Feature Per Cluster')
plt.xlabel('Feature')
plt.ylabel('Mean Value (Scaled)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

#### Interpretation of 2-Cluster Analysis
- Cluster 0 (Blue bars):
   - High: Customers in this cluster have a higher 'recency' value, meaning they made their last purchase more recently than customers in Cluster 1.
   - Lower Total Trips, Total Spend, Avg Spend, Avg Surge, Avg Tip: These customers tend to have fewer trips, lower total spending, and lower average spending, surge, and tip amounts.
   - Lower Active Days: This cluster shows a lower 'active_days' value, suggesting less frequent engagement over time.

Interpretation: Cluster 0 appears to represent Newer, Less Engaged, or Lower Value Customers. They might be new users who haven't accumulated many trips or high spending yet, or they could be less active users overall.

- Cluster 1 (Green bars):
   - Lower Recency: Customers in this cluster have a lower 'recency' value, indicating it's been longer since their last purchase compared to Cluster 0.
   - Higher Total Trips, Total Spend, Avg Spend, Avg Surge, Avg Tip: These customers tend to have more trips, higher total spending, and higher average spending, surge, and tip amounts.
   - Higher Active Days: This cluster has a higher 'active_days' value, suggesting more consistent engagement over time.

Interpretation: Cluster 1 seems to represent Older, More Engaged, or Higher Value Customers. These are likely loyal customers who use the service frequently and contribute more to revenue, even if their last activity was not as recent as Cluster 0.

#### Summary of Customer Segmentation
- Cluster 0 (Newer, Less Engaged, or Lower Value Customers): This cluster is characterized by higher recency (meaning more recent activity), lower total trips, lower total spend, lower average spend, lower average surge, lower average tip, and fewer active days.
- Cluster 1 (Older, More Engaged, or Higher Value Customers): This cluster is characterized by lower recency (meaning longer since last activity), higher total trips, higher total spend, higher average spend, higher average surge, higher average tip, and more active days.


Business Implications or Next Steps
- **Targeted Engagement for Cluster 0**: Develop engagement campaigns specifically for Cluster 0 customers (Newer, Less Engaged) to encourage more frequent usage and increase their overall value.
- **Loyalty & Retention for Cluster 1**: Implement loyalty programs or exclusive offers for Cluster 1 customers (Older, More Engaged) to ensure continued satisfaction and retention of these high-value users.
